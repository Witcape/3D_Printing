```python
import matplotlib.pyplot as plt
import numpy as np

# Data for accuracy
models = ['Normal', 'SRGAN', 'CLAHE', 'SRCNN']
accuracy_final_unet = [38.76, 34.01, 39.79, 35.92]
accuracy_final_cropped_unet = [53.91, 50.39, 50.78, 56.25]

# X-axis positions
x = np.arange(len(models))

# Plotting the accuracy comparison in a single graph
plt.figure(figsize=(10, 6))
plt.bar(x - 0.2, accuracy_final_unet, 0.4, label='Final_UNET', color='#1f77b4')  # Blue
plt.bar(x + 0.2, accuracy_final_cropped_unet, 0.4, label='Final_Cropped_UNET', color='#ff7f0e')  # Orange

# Adding labels and title
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison: Final_UNET vs. Final_Cropped_UNET')
plt.xticks(x, models)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Graph for Final_UNET Dataset
plt.figure(figsize=(10, 6))
plt.bar(x, accuracy_final_unet, color='#1f77b4', width=0.5)
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison for Final_UNET Dataset')
plt.xticks(x, models)
plt.tight_layout()
plt.show()

# Graph for Final_Cropped_UNET Dataset
plt.figure(figsize=(10, 6))
plt.bar(x, accuracy_final_cropped_unet, color='#ff7f0e', width=0.5)
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison for Final_Cropped_UNET Dataset')
plt.xticks(x, models)
plt.tight_layout()
plt.show()
```
