```python
import matplotlib.pyplot as plt
import numpy as np

# Data for accuracy
models = ['Final_VIT', 'SRGAN', 'CLAHE', 'SRCNN']
accuracy_normal = [87.34, 86.15, 86.30, 87.66]
accuracy_cropped = [96.09, 95.35, 96.88, 94.16]

# X-axis positions
x = np.arange(len(models))

# Improved color scheme
plt.figure(figsize=(10, 6))
plt.bar(x - 0.2, accuracy_normal, 0.4, label='Normal', color='#1f77b4')  # Blue
plt.bar(x + 0.2, accuracy_cropped, 0.4, label='Final_Cropped_VIT', color='#ff7f0e')  # Orange

# Adding labels and title
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison among Different Models and Preprocessing Techniques')
plt.xticks(x, models)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for accuracy
models = ['Normal', 'SRGAN', 'CLAHE', 'SRCNN']
accuracy_final_vit = [87.34, 86.15, 86.30, 87.66]
accuracy_final_cropped_vit = [96.09, 95.35, 96.88, 94.16]

# X-axis positions
x = np.arange(len(models))

# Graph for Final_VIT Dataset
plt.figure(figsize=(10, 6))
plt.bar(x, accuracy_final_vit, color='#1f77b4', width=0.5)
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison for Final_VIT Dataset')
plt.xticks(x, models)
plt.tight_layout()
plt.show()

# Graph for Final_Cropped_VIT Dataset
plt.figure(figsize=(10, 6))
plt.bar(x, accuracy_final_cropped_vit, color='#ff7f0e', width=0.5)
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison for Final_Cropped_VIT Dataset')
plt.xticks(x, models)
plt.tight_layout()
plt.show()
```
