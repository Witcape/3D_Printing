```python
import matplotlib.pyplot as plt
import numpy as np

# Data for accuracy
models = ['Normal', 'SRGAN', 'CLAHE', 'SRCNN']
accuracy_final_deit = [93.80, 90.93, 91.47, 92.44]
accuracy_final_cropped_deit = [96.88, 96.90, 97.66, 94.89]

# X-axis positions
x = np.arange(len(models))

# Plotting the accuracy comparison in a single graph
plt.figure(figsize=(10, 6))
plt.bar(x - 0.2, accuracy_final_deit, 0.4, label='Final_Deit', color='#1f77b4')  # Blue
plt.bar(x + 0.2, accuracy_final_cropped_deit, 0.4, label='Final_Cropped_Deit', color='#ff7f0e')  # Orange

# Adding labels and title
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison: Final_Deit vs. Final_Cropped_Deit')
plt.xticks(x, models)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Graph for Final_Deit Dataset
plt.figure(figsize=(10, 6))
plt.bar(x, accuracy_final_deit, color='#1f77b4', width=0.5)
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison for Final_Deit Dataset')
plt.xticks(x, models)
plt.tight_layout()
plt.show()

# Graph for Final_Cropped_Deit Dataset
plt.figure(figsize=(10, 6))
plt.bar(x, accuracy_final_cropped_deit, color='#ff7f0e', width=0.5)
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison for Final_Cropped_Deit Dataset')
plt.xticks(x, models)
plt.tight_layout()
plt.show()
```
