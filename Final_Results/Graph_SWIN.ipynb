```python
import matplotlib.pyplot as plt
import numpy as np

# Data for accuracy
models = ['Normal', 'SRGAN', 'CLAHE', 'SRCNN']
accuracy_final_swin = [25.84, 37.03, 38.50, 34.51]
accuracy_final_cropped_swin = [32.81, 32.56, 32.81, 35.04]

# X-axis positions
x = np.arange(len(models))

# Plotting the accuracy comparison in a single graph
plt.figure(figsize=(10, 6))
plt.bar(x - 0.2, accuracy_final_swin, 0.4, label='Final_SWIN', color='#1f77b4')  # Blue
plt.bar(x + 0.2, accuracy_final_cropped_swin, 0.4, label='Final_Cropped_SWIN', color='#ff7f0e')  # Orange

# Adding labels and title
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison: Final_SWIN vs. Final_Cropped_SWIN')
plt.xticks(x, models)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Graph for Final_SWIN Dataset
plt.figure(figsize=(10, 6))
plt.bar(x, accuracy_final_swin, color='#1f77b4', width=0.5)
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison for Final_SWIN Dataset')
plt.xticks(x, models)
plt.tight_layout()
plt.show()

# Graph for Final_Cropped_SWIN Dataset
plt.figure(figsize=(10, 6))
plt.bar(x, accuracy_final_cropped_swin, color='#ff7f0e', width=0.5)
plt.xlabel('Preprocessing Techniques')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison for Final_Cropped_SWIN Dataset')
plt.xticks(x, models)
plt.tight_layout()
plt.show()
```
